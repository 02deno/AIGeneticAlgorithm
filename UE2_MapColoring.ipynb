```python
# Komşu Matriksi oluşturma
# diagonal = 0 olmalı
# simetrik olmalı
def create_graph(n) : 
    graph = []
    for i in range(n):
        vertex = []
        for j in range(n):
            vertex.append(random.randint(0, 1))
        graph.append(vertex)
    for i in range(n):
        for j in range(0, i):
            graph[i][j] = graph[j][i] # simetrik
    for i in range(n):
        graph[i][i] = 0 # diogonal = 0
    for v in graph:
        print(v)
    return graph    
```

```python
# Find the upper bound for coloring
def max_color(graph) :
    max_num_colors = 1
    for i in range(n):
        if sum(graph[i]) > max_num_colors:
            max_num_colors = sum(graph[i]) + 1
    print("Max Number of Colors:",max_num_colors)
    return max_num_colors
```

```python
	'''Fitness'''
	def fitness(graph, individual):
		fitness = 0
		for i in range(n):
			for j in range(i, n):
				if(individual[i] == individual[j] and graph[i][j] == 1):
					fitness += 1
		return fitness
```

```python
	'''Tournament Selection'''
	def tournament_selection(population):
		new_population = []
		for j in range(2): # 2 parents
			random.shuffle(population)
			for i in range(0, len(population)-1, 2):
				if fitness(graph, population[i]) < fitness(graph, population[i+1]):
					new_population.append(population[i])
				else:
					new_population.append(population[i+1])
		return new_population
```

```python
	'''Crossover'''
	def crossover(parent1, parent2):
		position = random.randint(2, n-2) # n = countries,crossover point 
		child1 = []
		child2 = []
		for i in range(position+1):
			child1.append(parent1[i])
			child2.append(parent2[i])
		for i in range(position+1, n):
			child1.append(parent2[i])
			child2.append(parent1[i])
		return child1, child2
```

```python
	def mutation(individual,probability):
		check = random.uniform(0, 1)
		if(check <= probability):
			position = random.randint(0, n-1)
			individual[position] = random.randint(1, number_of_colors)
		return individual
```

```python
def create_population(population_size) :
    #Create Population
    generation = 0
    population = []
    for i in range(population_size):
        individual = create_individual()
        population.append(individual)
    return population
```

```python
def find_fittest(graph,population):
	best_fitness = fitness(graph, population[0])
	fittest_individual = population[0]
	for individual in population:
		if(fitness(graph, individual) < best_fitness):
			best_fitness = fitness(graph, individual)
			fittest_individual = individual
	return best_fitness,fittest_individual
```

```python
def check_for_solution(best_fitness,number_of_colors):
	solution = True
	if(best_fitness != 0):
		solution = False
		print("Graph is ", number_of_colors+1, " colorable")
	else:
		number_of_colors -= 1
	return solution,number_of_colors
```
